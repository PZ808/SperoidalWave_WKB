# Residue calculus

The two de Rham cohomology classes of the genus 1 curve $f=0$ can be represented as $z^4,1$. 
Let $\Omega = z^4(d\frac{x}{z} \wedge d \frac{y}{z^2})$ .
The polynomials are mapped to differential forms by 
$$ z^4\mapsto \text{res} \frac{z^4 \Omega}{f^2}, \,\,\,\, \text{ and } \,\,\,\,\,\,\, 1 \mapsto \text{res} \frac{\Omega}{f} $$
Note the invariance under rescaling $[x,y,z] \mapsto [\lambda x,\lambda^2 y,\lambda z]$.
The residue map $\text{res}$ is defined in local coordinates ($z=1$) through the following: 
Let $\eta$ be a 1-form regular on $f=0$. 
Then
$$\text{res } \eta \wedge \frac{df}{f} = \eta\vert_{f=0} $$
Note, that the residue of a form with pole order not equal to 1 by first reducing the pole order to 1 modulo exact forms using the `Pole Reduction' formula:
$$ d \frac{\eta}{f^m} = \frac{d \eta}{f^m} + \frac{1}{m} \frac{\eta \wedge df}{f^{m+1}} $$

Note the special case where $f = \frac{1}{2} y^2 +V(x) - E$.
The pole reduction formula implies that (assume $m>n$) 
$$\tag{1}
  g(x)  \frac{ y^{2n}  dx \wedge dy }{f^m} =  g(x)  \frac{ y^{2n-1}  dx \wedge df }{f^m} = \frac{2n-1}{m-1}  g(x)  \frac{ y^{2n-2}  dx \wedge df }{f^{m-1}} = \dots = \frac{2^n \Gamma \left(n+\frac{1}{2}\right) \Gamma (m-n+1)}{\sqrt{\pi } \Gamma (m)} \frac{g(x)dx\wedge dy}{f^{m-n}}
$$

Using the residue formula the classical action $pdx \vert_{H=E}$ is the residue of
$$ \frac{pdx\wedge dH }{H-E}  $$
which in the $[x,y,z]$ coordinates is given by 
$$\sigma = \frac{y^2 \Omega }{(z^2-x^2)z^2 f} $$
There are poles at $x=+1,-1,\infty $.

# Reduction of Pole order 

Below is code which implements the Griffiths-Dwork reduction

```sage
def NormalForm(x,J):            # Compute the quotients and remainders of x mod the ideal J
    r = x.reduce(J)             # computes the remainder r using `reduce'
    l=(x-r).lift(J)             # computes the quotients l by removing the remainder and using the `lift' command from sage
    return [r,l]

def PoleReduce(polynom, J):                                                 # implements the reduction of pole order 
    qr = NormalForm(polynom,J)                                              # find the quotients and remainders of polynom w.r.t. the ideal J
    xs = J.ring().gens()                                                    # find the generators of the ring in which J lives, i.e. the variables x,y,z
    q = sum([diff(qr[1][i], xs[i]) for i in range(0,len(xs)) ])             # finds the derivatives of the quotients 
    hcs = q.homogeneous_components()                                        # Divides the derivatives of the quotients into their homogenous components, a necessary step for dividing by the pole order of each term
    wdim = sum([ xx.degree() for xx in xs])                                 # calculates the total weighted degree of variables
    degf = J.gens()[0].degree()+ xs[0].degree()                             # calculates the degree of the polynomial f whose jacobian ideal is J in the weighted projective space
    out = 0                                                                 # initialize the following for loop
    for h in sorted(hcs):                                                   # loop through the homogenous components of q
        # infer the pole order m from the degree using h=m*deg(f)-w*dim
        out = out + hcs[h]/((h+wdim)/degf)                                  # and divide by the pole order determined by the Reduction-of-Pole-order formula
    return out+qr[0] 

def PoleReduceFull(polynom, J):                                             # Applies PoleReduceFull the maximum number of times needed to completely reduce the polynomial according to the Griffiths-Dwork Reduction    
    xs = J.ring().gens()                                                    # find the generators of the ring in which J lives, i.e. the variables
    wdim = sum([ xx.degree() for xx in xs])                                 # calculates the total weighted degree of variables
    degf = J.gens()[0].degree()+ xs[0].degree()                             # calculates the degree of the polynomial f whose jacobian ideal is J in the weighted projective space
    maxPoleOrder = (max(polynom.homogeneous_components()) + wdim)/degf      # calculates the maximum pole order of the input polynomial
    out = PoleReduce(polynom,J)                                             # initialize the following loop by applying PoleReduce once
    for i in range(maxPoleOrder):                                           # apply pole reduce a time equal to the max pole order 
        out = PoleReduce(out,J)
    return out
```

# Quantum corrections

Goal: Compute the decomposition of the quantum corrections 
$$N_{\gamma} = N_{0,\gamma} + \hbar^2 N_{2,\gamma} + \dots $$

First generate differentials $\tau_k$ such that $\partial_E N_{\gamma} = 2\pi i \int_{T(\gamma)} \tau_{k}$.
They can be generated with the following linear recurrence relation:
$$ \tau_{k} = \frac{i p}{H-E} \partial_q \tau_{k-1} + \frac{1}{2} \frac{1}{H-E} \partial_q^2 \tau_{k-2}$$
with initial conditions
$$\tau_0 = \frac{dx\wedge dy}{H-E} \,\,\,\,\,\,\,\,\,\,\, \text{ and }, \,\,\,\,\,\,\,\,\,\,\, \tau_{-1} =0.$$

First: Transform the operators $g_1 = \frac{i p}{H-E} \partial_q$ and $g_2 = \frac{1}{2} \frac{1}{H-E} \partial_q^2$ to the coordinate system of $\mathbf{P}^{(1,2,1)}$

Note that $x=\sin(q)$ and $y= \frac{y}{\cos q}$

Which means that
$\frac{\partial}{\partial q} = \frac{\partial q}{\partial x}\frac{\partial }{\partial x} + \frac{\partial q}{\partial y}\frac{\partial }{\partial y}  =... $

The recurrence mapped onto the numerators of rational differentials is implemented below.

```sage
S.<x,y,z> = PolynomialRing(F, order=TermOrder('wdegrevlex', (1,2,1)))

def mhat(U):
    return (2*y*diff(U,y)+x*diff(U,x)+z*diff(U,z)+4*U)/4

def delta__1(U):
    return z*(diff(U, x) - diff(f, x)*mhat(U))

def delta__2(U):
    return (diff(U, y) - diff(f, y)*mhat(U))

def g__1(U):
    return (y*((z^2 - x^2)*delta__1(U) - z*x*y*delta__2(U))*I).expand()

def g__2(U):
    return ((1/2*z^4 - z^2*x^2 + 1/2*x^4)*delta__1(delta__1(U)) + y*z*(-z^2*x + x^3)*delta__2(delta__1(U)) + 1/2*y^2*x^2*delta__2(z^2*delta__2(U)) + 1/2*z*(-z^2 + x^2)*(y*z*delta__2(U) + x*delta__1(U)) ).expand()


def Pgen(max=Infinity):
    n = 0
    an = 1
    bn=0
    ao = 0
    while n <= max:
        yield an
        n += 1  
        bn = ao
        ao = an
        an = (g__1(an) + g__2(bn)).expand()
```

To compute the action split the above polynomial into quotient and remainder when divided by $x^{2} z^{2} - z^{4}$

```sage
dE = S.ideal([S(-diff(symbolic_expression(f),E))])

def mhatInv(polynom, J):
    xs = J.ring().gens()
    hcs = polynom.homogeneous_components()
    wdim = sum([ xx.degree() for xx in xs])
    degf = J.gens()[0].degree()+ xs[0].degree()
    out = 0
    for h in sorted(hcs):
        out = out + hcs[h]/((h+wdim)/degf)
    return out


def IntdE(poly):
    return mhatInv(S(poly).lift(dE)[0],J)
```

## GD Reduction

We work with the Jacobian ideal
$$J = (f_x, f_y, f_z) $$
If $P=r+\sum_i q_i\partial_i f$ with $r$ normal form modulo $J$, then for the pole-$m$ form $P\Omega/f^m$
Griffith's says we can reduce
$$P\Omega/f^m \equiv r\Omega/f^m + \frac{1}{m-1}\frac{\Omega\sum_i \partial_i{q_i}}{f^{m-1}}$$
so algorithmically
- reduce gives $r$
- lift give the $q_i$
- sum(diff(q_i,x_i)) gives the divergence,
- then we continue recursively

```sage
F0 = PolynomialRing(QQ, 'mu0,mu1,mu2,E')
F0.inject_variables()
mu0,mu1,mu2,E = F0.gens()
K = F0.fraction_field()

S.<x,y,z> = PolynomialRing(K, order=TermOrder('wdegrevlex', (1,2,1)))


f  = 1/2*y^2 + x^4 - mu2*x^3*z + (E-2)*x^2*z^2 + (mu1+mu2)*x*z^3 + (mu0-E+1)*z^4
fx = diff(f,x)
fy = diff(f,y)
fz = diff(f,z)
fE = x^2*z^2-z^4

J = S.ideal([fx,fy,fz])
GB = J.groebner_basis()
```

```sage
def wmons(d):
    mons = []
    for a in range(d+1):
        for b in range(d//2 + 1):
            c = d - a - 2*b
            if c >= 0:
                mons.append(x^a * y^b * z^c)
    return mons

def coeffvec(p, mons):
    return vector(K, [p.monomial_coefficient(m) for m in mons])
```

```sage
def decompose(Q, d, keep=[], allow_f=True):
    mons = wmons(d)

    gens = []
    tags = []

    if allow_f and d >= 4:
        for m in wmons(d-4):
            gens.append(m*f)
            tags.append(("f", m))

    for m in wmons(d-3):
        gens.append(m*fx)
        tags.append(("fx", m))
        gens.append(m*fz)
        tags.append(("fz", m))

    for m in wmons(d-2):
        gens.append(m*fy)
        tags.append(("fy", m))

    for b in keep:
        gens.append(b)
        tags.append(("keep", b))

    M = matrix(K, [coeffvec(g, mons) for g in gens]).transpose()
    rhs = coeffvec(Q, mons)
    sol = M.solve_right(rhs)

    return sol, tags
```

```sage
def unpack_solution(sol, tags):
    Ax = S(0); Ay = S(0); Az = S(0); B = S(0)
    keep_coeffs = {}

    for c, tag in zip(sol, tags):
        kind, obj = tag
        if kind == "fx":
            Ax += c*obj
        elif kind == "fy":
            Ay += c*obj
        elif kind == "fz":
            Az += c*obj
        elif kind == "f":
            B += c*obj
        elif kind == "keep":
            keep_coeffs[obj] = c

    return Ax, Ay, Az, B, keep_coeffs
```

### Notation

$\Pi_A = \int_\gamma \omega_A$ and $\vec \Pi = (\Pi_0,\Pi_1)^T$
are the periods of the basis differentials
$\omega_0=\Omega/f$ and $\omega_1= z^4\Omega/f^4$.

### Polynomals $P$ and $P^{(2)}$

$P$ = numerator for $\partial_\lambda \omega_1 = -2z^4 f_\lambda$  [GM route]
$P^{(2)}$ = raw numerator for $\partial_\lambda^2\omega_0 = 2 (f_\lambda)^2$  [PF route]

### after pole reduction

P1 = pole reduced version of $P$ [GM route]
p1 = pole-reduced version of $P^{(2)}$ [PF route]

### after moding out ideal $J$

r1 numerator for $\partial_\lambda \omega_0$   : r1 = (-f_lambda).reduce(J)
r2 numerator for $\partial_\lambda \omega_1$   : r2 = P1.reduce(J)
p2 numerator for $\partial_\lambda^2 \omega_0$ : p2 = p1.reduce(J)

### coefficient extraction

Expand reduced numerators in a basis.
Fixed basis in the GM route : $\{ f, z^4 \}$
Adapted basis (PF scalar route) $\{f, r1 \}$

#### coefficients

From r1: $r1 = a_1 f + b_1 z^4$
$a_1$: coefficient of $f$ contributed to $\omega_0$
$b_1$: coefficient of $f$ conrtibuted to $\omega_1$
so $\partial_\lambda\omega_0 = a_1 \omega_0 + b_1 \omega_1$

From r2: $r2 = a_2 f + b_2 z^4$
$a_2$: coefficient of $f$ contributed to $\omega_0$
$b_2$: coefficient of $f$ conrtibuted to $\omega_1$
so $\partial_\lambda\omega_1 = a_2 \omega_0 + b_2 \omega_1$

From p2 (direct PF route):
$p2 = Af + B r1 $
$A$: coefficient of $\omega_0$
$B$: coefficient of $\omega_1$

So the GM matrix is

$$\partial_\lambda \vec\omega = A_\lambda \vec\omega $$


$$
A_\lambda =
\begin{pmatrix}
a_1 & b_1 \\
a_2 & b_2
\end{pmatrix}.
$$


Special simplifications $\partial_{\mu_0}f = z^4$  so $r1 = -z^4$ , $a_1  = 0$ and $b_1  = -1$.   Then $\Omega_{\mu_0\mu_0} = -b_2 \Omega_{\mu_0} +a_2 \Omega = 0$.

```sage
def NormalFormOpted(p, J):
    """
    Return remainder r and coefficient vector qs such that
    p = r + sum(qs[i] * J.gens()[i]).
    """
    r = p.reduce(J)
    qs = (p - r).lift(J)
    return r, qs


def PoleReduceOpted(p, J, degf, weights):
    """
    One Griffiths-Dwork pole reduction step.

    Input:
      p       numerator polynomial
      J       Jacobian ideal (fx, fy, fz)
      degf    weighted degree of f
      weights list/tuple of variable weights, e.g. (1,2,1)

    Output:
      reduced numerator after one step
    """
    r, qs = NormalFormOpted(p, J)
    xs = J.ring().gens()

    divq = sum(qs[i].derivative(xs[i]) for i in range(len(xs)))
    hcs = divq.homogeneous_components()

    wdim = sum(weights)
    out = r.parent()(0)

    for h in sorted(hcs):
        m = ZZ((h + wdim) / degf)
        # infer the pole order m from the degree using h=m*deg(f)-w*dim
        out += hcs[h] / (m - 1)

    return out

def PoleReduceFull(p, J, degf, weights):
    """
    Iterate Griffiths-Dwork reduction until no further pole lowering is possible.
    """
    wdim = sum(weights)
    hmax = max(p.homogeneous_components())
    maxPoleOrder = ZZ((hmax + wdim) / degf)

    out = p
    for _ in range(maxPoleOrder - 1):
        newout = PoleReduceOpted(out, J, degf, weights)
        if newout == out:
            break
        out = newout

    return out
```

## GM connection
Compute the Gaussâ€“Manin matrix $A_\lambda$ in the basis
 $\omega_0 = \operatorname{Res}(\Omega/f)$, $\omega_1 = \operatorname{Res}(z^4 \Omega / f^2)$.

### First row: derivative of $\omega_0$ ---

We start from
$$\partial_\lambda \omega_0 = - \operatorname{Res}\frac{f_\lambda \Omega}{f^2},
\qquad f_\lambda = \partial_\lambda f. $$

The numerator is the degree-4 polynomial
$$P^{(0)}_\lambda = -f_\lambda.$$

Reducing modulo the Jacobian ideal $J=(f_x,f_y,f_z)$ gives
$$r^{(0)}_\lambda = (-f_\lambda)\bmod J = c_\lambda z^4 + d_\lambda f.$$

Since
$$\frac{f\Omega}{f^2} = \frac{\Omega}{f},$$
the $f$-term contributes to $\omega_0$, while the $z^4$ term contributes to $\omega_1$. Therefore
$$\partial_\lambda \omega_0 = a_1 \omega_0 + b_1 \omega_1.$$

So the first row of $A_\lambda$ is $(a_1, b_1 )$.

### Second row: derivative of $\omega_1$

We have
$$\partial_\lambda \omega_1
= \partial_\lambda \operatorname{Res}\frac{z^4 \Omega}{f^2}
= -2 \operatorname{Res}\frac{f_\lambda z^4 \Omega}{f^3}.$$

Define the degree-8 numerator
$$P^{(1)}_\lambda = -2 f_\lambda z^4.$$

After one Griffiths reduction step (pole reduction), this becomes a degree-4 polynomial
$$P^{(1)}_{\lambda,\mathrm{red}}.$$

Reducing modulo $J$ gives
$$r^{(1)}_\lambda = a_2 f + b_2 z^4.$$

Hence
$$\partial_\lambda \omega_1 = a_2 \omega_0 + b_2 \omega_1,$$

so the second row of $A_\lambda$ is $(\beta_\lambda, \alpha_\lambda)$.

### Gaussâ€“Manin matrix

$$
A_\lambda =
\begin{pmatrix}
a_{1\lambda} & b_{1\lambda} \\
a_{2\lambda} & b_{2\lambda}
\end{pmatrix}.
$$


For $\lambda = \mu_0$:
$$f_{\mu_0} = z^4 \quad \Rightarrow \quad \partial_{\mu_0} \omega_0 = -\omega_1,$$
so the first row is $(0,-1)$. The second row comes from reducing
$$P^{(1)}_{\mu_0} = -2 z^8.$$

For $\lambda = \mu_1$:
$$f_{\mu_1} = x z^3.$$
The first row comes from reducing
$$-x z^3 \equiv b_{1\mu_1} z^4 + a_{1\mu_1} f \quad (\mathrm{mod}\,J),$$
and the second row from reducing
$$P^{(1)}_{\mu_1} = -2 x z^7.$$

 ---  dictionary ---

$$r1 = r^{(0)}_\lambda \quad \longrightarrow \quad (a_1, b_1),$$

$$r2 = r^{(1)}_\lambda \quad \longrightarrow \quad (a_2, b_2).$$

```sage
# weighted monomials in weights (1,2,1)
def wmons2(d, R):
    x,y,z = R.gens()
    out = []
    for a in range(d+1):
        for b in range(d//2 + 1):
            for c in range(d+1):
                if a + 2*b + c == d:
                    out.append(x^a * y^b * z^c)
    return out

def coeffvec2(p, mons):
    return vector(p.parent().base_ring(), [p.monomial_coefficient(m) for m in mons])
```

```sage
def decompose_deg4(r, f, fx, fy, fz, mons_deg1, mons_deg2):
    """
    Write degree-4 polynomial r as

        r = a*z^4 + b*f + qx*fx + qy*fy + qz*fz

    where qx,qz have weighted degree 1 and qy has weighted degree 2.
    """

    R = r.parent()
    x,y,z = R.gens()

    # generic degree-1 and degree-2 ansatz
    qx = sum(var('ax%s'%i)*m for i,m in enumerate(mons_deg1))
    qz = sum(var('az%s'%i)*m for i,m in enumerate(mons_deg1))
    qy = sum(var('ay%s'%i)*m for i,m in enumerate(mons_deg2))

    a = var('a')
    b = var('b')

    expr = expand(r - (a*z^4 + b*f + qx*fx + qy*fy + qz*fz))
    eqs = [expr.monomial_coefficient(m) == 0 for m in expr.monomials()]
    sol = solve(eqs, [a,b] + qx.variables() + qy.variables() + qz.variables(), solution_dict=True)

    return sol[0][a], sol[0][b]
```


Decompose a degree-4 numerator against $\{f,b_1\}$ modulo the ideal $J$. This solves
$$  p = A f + B b_1 + q_xf_x + q_y f_y + q_q f_z $$

```sage
def coeffs_in_basis(p, b1, f, fx, fy, fz):
    """
    Decompose a weighted-degree-4 polynomial p as

        p = A*f + B*b1 + qx*fx + qy*fy + qz*fz

    where deg(qx)=deg(qz)=1 and deg(qy)=2
    in weights (1,2,1).

    Returns:
        A, B, qx, qy, qz
    """
    R = p.parent()
    K = R.base_ring()
    x,y,z = R.gens()

    mons4 = wmons2(4, R)
    mons1 = wmons2(1, R)   # [x, z]
    mons2 = wmons2(2, R)   # [x^2, x*z, z^2, y]

    basis_polys = [f, b1] \
                  + [m*fx for m in mons1] \
                  + [m*fy for m in mons2] \
                  + [m*fz for m in mons1]

    M = matrix(K, len(mons4), len(basis_polys),
               lambda i,j: basis_polys[j].monomial_coefficient(mons4[i]))
    rhs = coeffvec2(p, mons4)

    # Solve one consistent solution
    sol = M.solve_right(rhs)

    A = sol[0]
    B = sol[1]

    off = 2
    qx = sum(sol[off+i] * mons1[i] for i in range(len(mons1)))
    off += len(mons1)

    qy = sum(sol[off+i] * mons2[i] for i in range(len(mons2)))
    off += len(mons2)

    qz = sum(sol[off+i] * mons1[i] for i in range(len(mons1)))

    # Optional uniqueness check for A,B:
    # if every kernel vector has first two entries zero, then A,B are unique.
    ker = M.right_kernel()
    for v in ker.basis():
        if v[0] != 0 or v[1] != 0:
            print("Warning: A,B are not uniquely fixed by this ansatz.")

    return A, B, qx, qy, qz
```

 Let $b_1$ be the degree-4 numerator representing $\partial_\lambda \omega_0$ and
  let $p_1$ be the degree-4 numerator representing $\partial_\lambda^2 \omega_0$

```sage
def PF_data_from_second_derivative(p2, b1, f, fx, fy, fz):
    """
    Given:
      b1 = numerator representing d_lambda omega0
      p2 = numerator representing d_lambda^2 omega0

    returns A,B such that
      d_lambda^2 omega0 = A*omega0 + B*d_lambda omega0

    hence PF operator:
      D_lambda^2 - B D_lambda - A
    """
    A, B, qx, qy, qz = coeffs_in_basis(p2, b1, f, fx, fy, fz)
    return A, B
```
